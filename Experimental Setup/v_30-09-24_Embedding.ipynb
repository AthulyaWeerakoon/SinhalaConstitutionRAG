Clean

```python
import re
from itertools import chain

def preprocess_text(text):
    # Split by period '.'
    sentences = text.split('.')

    processed_sentences = []
    for sentence in sentences:
        # Remove English letters
        sentence = re.sub(r'[a-zA-Z]', '', sentence)

        # Remove punctuation and special characters (retain Sinhala letters)
        sentence = re.sub(r'[^\u0D80-\u0DFF\s]', '', sentence)  # Unicode range for Sinhala is \u0D80 to \u0DFF

        # Remove extra whitespace
        sentence = re.sub(r'\s+', ' ', sentence).strip()

        if sentence:  # Add sentence if not empty after cleaning
            processed_sentences.append(sentence)

    # Return the processed text as a single string, joined by periods
    return processed_sentences

# Apply the preprocessing to the entire dataset
processed_texts = [preprocess_text(text) for text in texts]
print(processed_texts[:5])

# Flatten the list of lists into a 1-dimensional list of sentences
all_sentences = list(chain.from_iterable(processed_texts))
print(all_sentences[:10])
```

Tokenize

```python
import tensorflow as tf

# Setup vocab size
# vocab_size = 20000

# Initialize the tokenizer
tokenizer = tf.keras.preprocessing.text.Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(all_sentences)

# Convert sentences to sequences of integers
sequences = tokenizer.texts_to_sequences(all_sentences)
print(sequences[:5])

# Test
print(tokenizer.sequences_to_texts(sequences[:5]))
```

Set up MLM goal

```python
import random
import numpy as np

mask_token = "<MASK>"
sequence_length = 50

# Add the mask token to the tokenizer
mask_token_idx = len(tokenizer.word_index) + 1
tokenizer.word_index[mask_token] = mask_token_idx
tokenizer.index_word[mask_token_idx] = mask_token

# Update vocab_size to include the mask token
# vocab_size += 2  # +1 for OOV and +1 for MASK

# Pad sequences to the same length
padded_sequences = tf.keras.preprocessing.sequence.pad_sequences(sequences, maxlen=sequence_length, padding='post', truncating='post')
print(padded_sequences[:5])

# Function to create masked inputs and target outputs
def create_masked_inputs(sequences, mask_token_id, mask_prob=0.15):
    input_ids = []
    target_ids = []

    for seq in sequences:
        mask_prob_itter = mask_prob
        input_seq = []
        target_seq = []
        for token_id in seq:
            if token_id == 0:
                # Padding token, no prediction
                input_seq.append(0)
                target_seq.append(0)
            else:
                if random.random() < mask_prob_itter:
                    input_seq.append(mask_token_id)  # Replace with [MASK]
                    target_seq.append(token_id)      # Original token as target
                    mask_prob_itter -= 0.05
                else:
                    input_seq.append(token_id)       # Original token
                    target_seq.append(mask_token_id) # No prediction
                    mask_prob_itter += 0.05
        input_ids.append(input_seq)
        target_ids.append(target_seq)

    return np.array(input_ids), np.array(target_ids)

# Create masked inputs and targets
input_ids, target_ids = create_masked_inputs(padded_sequences, mask_token_idx)
print(input_ids[0])
print(target_ids[0])

# Test by converting sequences to words
print(tokenizer.sequences_to_texts(input_ids[:5]))
print(tokenizer.sequences_to_texts(target_ids[:5]))

# Print length
print(len(input_ids))
```

Define Loss Function to Exclude Masks

```python
@tf.keras.saving.register_keras_serializable('custom_loss')
def masked_loss_function(y_true, y_pred):
    """
    Calculates the masked loss for the MLM task.

    Args:
        y_true: Ground truth labels (tensor).
        y_pred: Predicted labels (tensor).

    Returns:
        The masked loss value (scalar).
    """
    # Ensure y_true has the correct shape for sparse_categorical_crossentropy
    y_true = tf.squeeze(y_true, axis=-1)  # Remove the extra dimension
    print(y_true, y_true.shape)

    # Calculate the loss
    loss = tf.keras.losses.sparse_categorical_crossentropy(y_true, y_pred)
    print(loss, loss.shape)

    # Create a mask to ignore padding tokens (represented by 0)
    mask = tf.cast(tf.math.not_equal(y_true, 0), dtype=loss.dtype)
    print(mask, mask.shape)

    # Apply the mask to the loss
    masked_loss = loss * mask
    print(masked_loss, masked_loss.shape)

    # Calculate the mean masked loss
    reduced_mean = tf.reduce_mean(masked_loss)
    print(reduced_mean)
    return reduced_mean
```

Custom Self Attention Layer

```python
@tf.keras.saving.register_keras_serializable('simple_self_attention')
class SelfAttention(tf.keras.layers.Layer):
    def __init__(self, units, **kwargs):
        super(SelfAttention, self).__init__(**kwargs)
        self.units = units

    def build(self, input_shape):
        self.Wq = self.add_weight(shape=(input_shape[-1], self.units), initializer='random_normal', trainable=True)
        self.Wk = self.add_weight(shape=(input_shape[-1], self.units), initializer='random_normal', trainable=True)
        self.Wv = self.add_weight(shape=(input_shape[-1], self.units), initializer='random_normal', trainable=True)

    def call(self, inputs):
        Q = tf.matmul(inputs, self.Wq)  # Query
        K = tf.matmul(inputs, self.Wk)  # Key
        V = tf.matmul(inputs, self.Wv)  # Value

        # Calculate attention scores
        scores = tf.matmul(Q, K, transpose_b=True) / tf.sqrt(float(self.units))
        attention_weights = tf.nn.softmax(scores, axis=-1)

        # Apply attention weights to the values
        output = tf.matmul(attention_weights, V)
        return output

    def get_config(self):
        config = super().get_config()
        config.update({"units": self.units})
        return config
```

Build Model (to Predict Next Word)

```python
def build_model(vocab_size, embedding_dim, lstm_units, sequence_length, dropout_rate):
    # Input layer
    input_layer = tf.keras.layers.Input(shape=(sequence_length,))  # Input shape based on tokenized sequences

    # Embedding layer
    embedding_layer = tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim)(input_layer)

    # Bidirectional LSTM layer
    lstm_out = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(lstm_units, return_sequences=True))(embedding_layer)

    # Custom Self-Attention layer
    attention_out = SelfAttention(units=lstm_units, name='self_attention_layer')(lstm_out)

    # Output layer to predict the masked token
    output = tf.keras.layers.Dense(vocab_size, activation='softmax', name='output')(attention_out)

    # Create the model
    model = tf.keras.models.Model(inputs=input_layer, outputs=output)

    # Compile the model
    model.compile(optimizer='adam', loss=masked_loss_function, metrics=['accuracy'])

    return model
```

Build model with hyperparameters

```python
# Hyperparameters
vocab_size = len(tokenizer.word_index) + 1        # Number of unique tokens in your vocabulary
embedding_dim = 256                               # Dimensionality of the embedding space
lstm_units = 128                                  # Number of units in LSTM layer
dropout_rate = 0.5                                # Dropout rate for regularization

# Initialize and summarize the model
mlm_model = build_model(vocab_size=vocab_size, embedding_dim=embedding_dim, lstm_units=lstm_units, sequence_length=sequence_length, dropout_rate=dropout_rate)
mlm_model.summary()
```

Train for MLM Objective

```python
from sklearn.model_selection import train_test_split

# Expand target_ids to have an extra dimension for sparse_categorical_crossentropy
target_ids_expanded = np.expand_dims(target_ids, axis=-1)

# Split the data
train_input, val_input, train_target, val_target = train_test_split(input_ids, target_ids_expanded, test_size=0.2, random_state=42)

print(np.squeeze(val_target[:5], axis=-1))
print(input_ids.shape)
print(target_ids_expanded.shape)
```

```python
import os

# Define callbacks
checkpoint_path = '/content/drive/MyDrive/LLM_Tasks/Task 1/Embedding Layer/Models/mlm_model.keras'
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
checkpoint = tf.keras.callbacks.ModelCheckpoint(checkpoint_path, monitor='val_loss', save_best_only=True)

# Restoring from checkpoint
# Check if the checkpoint exists
if os.path.exists(checkpoint_path):
    # Load the saved model architecture from the checkpoint
    custom_objects = {'masked_loss_function': masked_loss_function, 'SelfAttention': SelfAttention}
    saved_model = tf.keras.models.load_model(checkpoint_path, custom_objects=custom_objects)

    # Compare model structures by their configuration
    if mlm_model.get_config() == saved_model.get_config():
        print("Model architectures match. Loading weights from checkpoint.")
        mlm_model.load_weights(checkpoint_path)
    else:
        print("Model architectures do not match. Starting from scratch.")
else:
    print("No checkpoint found. Starting from scratch.")

# Train the model
history = mlm_model.fit(
    train_input,
    train_target,
    epochs=20,
    verbose=1,
    batch_size=32,
    validation_data=(val_input, val_target),
    callbacks=[early_stopping, checkpoint]
)
```

Save Dynamic Embedding Layer

```python
import os

# Extract dynamic embedding component
attention_embedding_model = tf.keras.models.Model(inputs=mlm_model.input, outputs=mlm_model.get_layer('self_attention_layer').output)

def save_model_with_incremental_name(model, base_filename='attention_embedding_model', extension='.keras'):
    # Initialize the version counter
    version = 0
    filename = f"{base_filename}{extension}"

    # Check if the file exists and keep incrementing the version number
    while os.path.exists(filename):
        version += 1
        filename = f"{base_filename}_{version}{extension}"

    # Save the model with the final filename
    model.save(filename)
    print(f"Model saved as {filename}")

# Example usage:
save_model_with_incremental_name(attention_embedding_model)
```
