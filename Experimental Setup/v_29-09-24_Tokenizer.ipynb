```python
from datasets import load_dataset
```

```python
dataset = load_dataset("0xAIT/SIDAC")
```

```python
dataset
```

```python
train_dataset = dataset["train"].remove_columns(['timestamp', 'url', 'source'])
```

```python
train_dataset = train_dataset.add_column('id', list(range(len(train_dataset))))
```

```python
dataset['train'] = train_dataset
```

```python
dataset ['train'][1]
```

```python
from tqdm.auto import tqdm

text_data = {i: [] for i in range(18)}

for sample in tqdm(dataset['train']):
    sample_id = sample['id']
    sample_text = sample['text'].replace('\n', '')


    file_number = sample_id % 18
    text_data[file_number].append(sample_text)

for file_count in range(18):
    with open(f'./data/text_{file_count}.txt', 'w', encoding='utf-8') as fp:
        fp.write('\n'.join(text_data[file_count]))
```

```python
from pathlib import Path
paths = [str(x) for x in Path('/content/data/').glob('**/*.txt')]
paths
```

```python
!pip install transformers
```

```python
from tokenizers import ByteLevelBPETokenizer
tokenizer = ByteLevelBPETokenizer()

tokenizer.train(files=paths, vocab_size=30000, min_frequency=2,
                special_tokens=['<s>', '<pad>',  '</s>',  '<unk>', '<mask>'])
```

```python
!mkdir SinhalaBERT
tokenizer.save_model("SinhalaBERT")
```
